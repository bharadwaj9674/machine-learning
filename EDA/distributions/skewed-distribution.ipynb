```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = pd.read_csv('income.csv')
```

```python
df = pd.DataFrame(file)
```

```python
df.head()
```

```python
plt.figure(figsize=(30, 10))
graph = sns.barplot(x=df['income($)'], y=df['count'], data=df)
```

```python
plt.figure(figsize=(30, 10))
graph = sns.barplot(x=df['income($)'], y=df['count'], data=df)
graph.set(xscale = "log")
plt.show()
```
