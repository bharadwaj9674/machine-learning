Correlation tests are performed to check how to features are depending on each other.

Correlation tests are generally performed for two numerical and continuous features.

If we get +1 correlation, then it indicates strong positive linear relationship and -1 indicates strong negative correlation.

0 to +1 indicate positive correlation, which means which increase in one value in the first feature, another value associated to that value in the second feature will also increase.

-1 to 0 indicate negative correlation, which means which increase in one value in the first feature, another value associated to that value in the second feature will decrease.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = pd.read_csv('Salary_dataset.csv')
```

```python
df = pd.DataFrame(file)
```

```python
df.shape
```

```python
df.head()
```

```python
df = df.drop('Unnamed: 0', axis = 1)
```

```python
sns.scatterplot(data = df, x = 'YearsExperience', y = 'Salary')
plt.show()
```

Pearsons correlation is used to measure the strength and direction of the two variable which are linearly associated.

Conditions to apply pearsons correlation:

- Two features are continuous
- Linear relation should be there between the features
- No outliers in the dataset

Pearsons correlation uses covariance to find the correlation between two variables.

```python
corr = df.corr(method='pearson')
sns.heatmap(corr, annot=True, cmap='Greens')
plt.show()
```

If the data is not linearly related we can use spearman correlation.

Spearman Correlation is used to measure the strength and direction of the two variables which are monotonically associated.

Monotonic means if one variable increase/decrease, other variable may increase/decrease. But the variables may not have a linear relationship.

Spearman correlation uses rank to find the correlation between two variables.

```python
corr = df.corr(method='spearman')
sns.heatmap(corr, annot=True, cmap='Greens')
plt.show()
```

Parametric is a statistical test which assumes parameters and the distributions about the dataset are known. It uses mean value to measure the central tendency.

eg: Linear Regression, Logistic Regression etc., are parametric models.

Non-Parametric is a statistical test which does not make any assumptions about the dataset. It measures the central tendency with the median.

eg: KNN, Decision trees etc., are non parametric models.



Kendall correlation(Non-Parametric) is an alternate to pearson's correlation(parametric), when the data we are working with has failed one or more assumptions of the persons correlation test then we can use Kendall correlation.

Kendall correlation measures the similarity of the orderings of the two variables.

Kendall correlations is also the best alternate to spearman correlation. When your sample size is large and if you are having many tied ranks then you can use kendall correlation. If your dataset is small then it may not give you accurate results.

Kendall correlation uses number of concordant and discordant pairs to find the correlation.

```python
corr = df.corr(method='kendall')
sns.heatmap(corr, annot=True, cmap='Greens')
plt.show()
```

```python
x = [-4,-3,-2,-1,0,1,2,3,4]
y = [-4,-3,-2,-1,0,1,2,3,4]

y_new = []

for i in y:
    y_new.append(i*i)
```

```python
df_1 = pd.DataFrame(list(zip(x,y)), columns = ['x','y'])
```

```python
sns.scatterplot(data = df_1, x = 'x', y = 'y')
plt.show()
```

```python
corr = df_1.corr(method='pearson')
sns.heatmap(corr, annot=True, cmap='Greens')
plt.show()
```

```python
corr = df_1.corr(method='spearman')
sns.heatmap(corr, annot=True, cmap='Greens')
plt.show()
```

```python
corr = df_1.corr(method='kendall')
sns.heatmap(corr, annot=True, cmap='Greens')
plt.show()
```

```python
df_2 = pd.DataFrame(list(zip(x,y_new)), columns = ['x','y_new'])
```

```python
sns.scatterplot(data = df_2, x = 'x', y = 'y_new')
plt.show()
```

```python
corr = df_2.corr(method='pearson')
sns.heatmap(corr, annot=True, cmap='Greens')
plt.show()
```
